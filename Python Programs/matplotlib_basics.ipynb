```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2])
b=np.array([9,7])
plt.plot(a,b)
plt.show()
```

```python
x=np.array([2,3])
y=np.array([4,5])

plt.plot(x,y,'*')
plt.show()
```

```python
x=np.array([1,2,3,4,5])
y=np.array([10,9,8,7,6])

plt.plot(x,y,'o')
plt.show()
```

```python
x=np.array([2,3,6,7])
y=np.array([10,5,8,2])

plt.plot(x,y,marker='*')
plt.show()
```
